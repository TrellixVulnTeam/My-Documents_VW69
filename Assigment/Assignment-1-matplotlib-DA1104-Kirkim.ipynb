## 1: Plot number of people by gender

```python
kill.head()
```

```python
fig, ax = plt.subplots() ###########################################################################################################

ax.bar(kill.gender.unique(),kill.gender.value_counts())

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.33, p.get_height()+10));
```

## 2: Plot 7 most common kill weapons

```python
kill.armed.value_counts().head(7)
```

```python
kill.armed.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(8,6))       

ax.bar(kill.armed.value_counts().head(7).index,kill.armed.value_counts().head(7))
plt.xticks(rotation=51)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+10));
```

## 3: Plot number of age of killed people under two groups : Under 25 and Above 25

```python

kill.head()
```

```python
kill['age_cat'] = np.where(kill['age'] > 25, 'above25', 'below25')
kill.age_cat.value_counts()
```

```python
fig, ax = plt.subplots()
ax.bar(kill['age_cat'].value_counts().index, kill['age_cat'].value_counts())

for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x()+0.33, p.get_height()+15));
```

```python
# with matplotlib
```
